**This notebook is to calculate the statistical metrics of the model.**

This will evaluate the model performance in emulating training dataset in different resolutions.
We used the mean absolute error, root mean square error, and r square as the statistical metrics.

```julia
## Import packages ##
#import Pkg
#Pkg.add("Plots")
using Plots
#Pkg.add("BenchmarkTools")
using BenchmarkTools
#Pkg.add("Flux")
using Flux
#Pkg.add("CSV")
using CSV
#Pkg.add("DelimitedFiles")
using DelimitedFiles
#Pkg.add("Statistics")
using Statistics
#Pkg.add("BSON")
using BSON: @save
using BSON: @load
#Pkg.add("StatsBase")
using StatsBase
```

```julia
## Evaluating the metrics of solvers in a resolution throughout 500 epcohs ##
## You need to change the resolution info (e.g., 8dx, 2dt case = ..._U_8x_2x... or ..._08x02t...)

temp = readdlm("Training_dataset/VL_GEOS_Jan_2019_NASA_GMAO_10_U_1x_4x.csv", ',', Float32)
xdim = size(temp, 1)
#500
for i in 1:500
    trial = readdlm("Train_outputs/01x04t/CNN_DLR_"*string(i)*"_01x04x_VL_GEOS_JAN_NASA_GMAO.csv", ',', Float32)
    
    dim = size(trial, 1)    
    nstep = Int(dim/xdim)
    answer = temp[:,1:nstep]
    
    mae = StatsBase.L1dist(trial, reshape(answer,dim))/(dim)
    rmse = StatsBase.L2dist(trial, reshape(answer,dim))/sqrt(dim)
    r2 = Statistics.cor(trial, reshape(answer, dim))^2
    
    println(i, "   ", mae, "   ", rmse, "   ", r2)
end
```

```julia
## For 2D demonstration

temp1 = readdlm("VL_GEOS_Jan_2019_NASA_GMAO_16x_64x.csv", ',', Float32)
xdim = size(temp1, 1)

for i in 1:500
    trial1 = readdlm("spacing_phys_norm_outputs/for2d/CNN_DLR_"*string(i)*"_16x64x_VL_GEOS_JAN_NASA_GMAO.csv", ',', Float32)
    
    dim1 = size(trial1, 1)
    nstep = Int(dim1/xdim)
    answer1 = temp1[:,1:nstep]
    
    mae1 = StatsBase.L1dist(trial1, reshape(answer1,dim1))/(dim1)
    rmse1 = StatsBase.L2dist(trial1, reshape(answer1,dim1))/sqrt(dim1)
    r2_1 = Statistics.cor(trial1, reshape(answer1, dim1))^2
    
    println(i, "   ", mae1, "   ", rmse1, "   ", r2_1)
end
```
