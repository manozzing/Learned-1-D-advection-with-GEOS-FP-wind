```julia
# Horizontal 2-D velocity field sampling
#2880
U_Jan = zeros(Float32, 201, 224, 2880)
V_Jan = zeros(Float32, 201, 224, 2880)

for i in 1:2880
    i_day = round(Int, i/288, RoundDown)+1
    i_hr = round(Int, (i/288 - i_day + 1)*24, RoundDown)
    i_min = round(Int, ((i/288 - i_day + 1)*24-i_hr)*60)
    if i_min == 60
        i_hr = i_hr + 1
        i_min = 0
    end
     
    if i_day < 10
        if i_hr < 10
            if i_min < 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *"0"*string(i_day)*"_"*"0"*string(i_hr)*"0"*string(i_min)*"z.nc4")
            elseif i_min >= 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *"0"*string(i_day)*"_"*"0"*string(i_hr)*string(i_min)*"z.nc4")
            end
        elseif i_hr >= 10
            if i_min < 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *"0"*string(i_day)*"_"*string(i_hr)*"0"*string(i_min)*"z.nc4")
            elseif i_min >= 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *'0'*string(i_day)*'_'*string(i_hr)*string(i_min)*"z.nc4")
            end
        end
    elseif i_day >= 10
        if i_hr < 10
            if i_min < 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *string(i_day)*'_'*'0'*string(i_hr)*'0'*string(i_min)*"z.nc4")
            elseif i_min >= 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *string(i_day)*'_'*'0'*string(i_hr)*string(i_min)*"z.nc4")
            end
        elseif i_hr >= 10
            if i_min < 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *string(i_day)*'_'*string(i_hr)*'0'*string(i_min)*"z.nc4")
            elseif i_min >= 10
                ds_NA = NCDataset("../GEOSChem/rundirs13.3.2/gcfptt_nested_NA_advect_only/OutputDir/GEOSChem.StateMet.2019"*"01"
                                        *string(i_day)*'_'*string(i_hr)*string(i_min)*"z.nc4")
            end
        end
    end
                                
    U_Jan[:,:,i] = ds_NA["Met_U"][:,:,1,1]'
    V_Jan[:,:,i] = ds_NA["Met_V"][:,:,1,1]'
end
```
